```python
import matplotlib.pyplot as plt
import matplotlib.axis as ax
%matplotlib inline
import pandas as pd
import numpy as np
from PIL import Image
```

```python
url = 'https://github.com/eueung/pilrek/raw/master/pilrek-anon.csv'
df  = pd.read_csv(url)
df
```

```python
df[df['CaRek Pilihan'] == 'Kadarsah Suryadi (FTI-TI)'].tail()
```

```python
Rektor1 = df[df['CaRek Pilihan'] == 'Kadarsah Suryadi (FTI-TI)']
```

```python
kat1 = Rektor1['Kategori Anda']
kat1
```

```python
kat1.value_counts()
```

```python
gra1 = kat1.value_counts().plot(kind='bar')
gra1
plt.savefig('gra1.png')
```
